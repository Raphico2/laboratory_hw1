```python
def compute_line_and_label(df):
  sequential_count = 0
  sepsis = 0
  row = 0
  column_label = df['SepsisLabel']
  for index, value in enumerate(column_label):
    if value == 1:
      sequential_count += 1
      if sequential_count == 6:
         sepsis = 1
         row = index - 5
    else:
      sequential_count = 0
    row = index
  return sepsis, row
```

```python
sepsis, row = compute_line_and_label(df)
print(sepsis, row)
```

```python
def compute_patient_data(df):
  data_line = []
  sepsis, row = compute_line_and_label(df)
  df = df.loc[0:row]
  line = []
  #for the clinical values we take the mean, max, min and std 
  for col in clinical_values:
    description = df[col].describe().to_dict()
    try:
       line.append(float(description['mean']))
    except:
       line.append('NaN')
    
    if description['count'] == 1:
      line.append(0)
    else:
      line.append(float(description['std']))
    try:
       line.append(float(description['max']))
    except:
       line.append('NaN')
      
    try:
       line.append(float(description['min']))
    except:
       line.append('NaN')
  
  for col in lab_variables:
    description = df[col].describe().to_dict()
    try:
       line.append(float(description['mean']))
    except:
       line.append(float(description['max']))

    if description['count'] == 1:
      line.append(0)
    else:
      line.append(float(description['std']))
    
    try:
       line.append(float(description['max']))
    except:
       line.append('NaN')
    
    try:
       line.append(float(description['min']))
    except:
       line.append('NaN')
  
  for col in demographical_values:
    description = df[col].describe().to_dict()
    try:
       line.append(int(description['mean']))
    except:
       line.append('NaN')
  
  line.append(int(sepsis))
  return line
```

Student Test for measuring the importance of each features with the label

```python
from scipy import stats

def student_t_test(dataframe, label_column):
    # Séparation des groupes en fonction de la colonne "label"
    group_0 = dataframe[dataframe[label_column] == 0]
    group_1 = dataframe[dataframe[label_column] == 1]

    # Parcours de chaque colonne du DataFrame
    for column in dataframe.columns:
        if column != label_column:
            data_group_0 = group_0[column]
            data_group_1 = group_1[column]

            # Test de Student indépendant
            t_statistic, p_value = stats.ttest_ind(data_group_0, data_group_1)

            print(f"Colonne : {column}")
            print(f"T-statistic : {t_statistic}")
            print(f"P-value : {p_value}\n")

student_t_test(train_df, 'label')
```

Correlation between each columns

```python
def calculate_correlations(dataframe, alpha):
    columns = dataframe.columns
    num_columns = len(columns)
    
    for i in range(num_columns):
        for j in range(i+1, num_columns):
            col1 = columns[i]
            col2 = columns[j]
            correlation = dataframe[col1].corr(dataframe[col2])
            
            if abs(correlation) > alpha:
                print(f"Correlation between {col1} and {col2} : {correlation}")

calculate_correlations(train_df, 0.7)
```

```python
from sklearn.utils import resample

def balance_dataset(dataframe, label_column):
    counts = dataframe[label_column].value_counts()
    minority_label = counts.idxmin()
    majority_label = counts.idxmax()
    minority_count = counts[minority_label]
    majority_count = counts[majority_label]

    print('balance of the dataset: ' + str(majority_count*100/(majority_count+minority_count)) + '% vs ' + str(minority_count*100/(majority_count+minority_count)) + '%' )

    minority_samples = dataframe[dataframe[label_column] == minority_label]
    majority_samples = dataframe[dataframe[label_column] == majority_label]
    minority_samples_upsampled = resample(minority_samples, replace=True, n_samples=majority_count, random_state=42)

    balanced_data = pd.concat([majority_samples, minority_samples_upsampled])
    counts = balanced_data[label_column].value_counts()
    minority_label = counts.idxmin()
    majority_label = counts.idxmax()
    minority_count = counts[minority_label]
    majority_count = counts[majority_label]

    print('balance of the dataset: ' + str(majority_count*100/(majority_count+minority_count)) + '% vs ' + str(minority_count*100/(majority_count+minority_count)) + '%' )

    return balanced_data
```
